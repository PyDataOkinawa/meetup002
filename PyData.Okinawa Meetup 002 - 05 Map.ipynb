```python
import folium
```

```python
from IPython.display import HTML
```

```python
def inline_map(m):
    # 中間生成される json が必要なプロットがあるため、一度 html として書き出し
    m.create_map(path='tmp.html')
    iframe = '<iframe srcdoc=\"{srcdoc}\" style=\"width: 100%; height: 400px; border: none\"></iframe>'
    return HTML(iframe.format(srcdoc=m.HTML.replace('\"', '&quot;')))
```

```python
m = folium.Map(location=[26.273, 127.729], zoom_start=17)
m.simple_marker([26.273947, 127.729032], popup='Gwave Cafe')
inline_map(m)
```

```python
m = folium.Map(location=[26.273, 127.729], zoom_start=10)
m.simple_marker([26.273947, 127.729032], popup='Gwave Cafe')
inline_map(m)
```

```python
topo_path = r'okinawa.json'
```

```python
tiles = ['Mapquest Open', 'Open Street Map', 'Mapbox Bright', 'Stamen Toner', 'Stamen Terrain']
```

```python
m = folium.Map(location=[26.273, 127.729], tiles=tiles[0], zoom_start=10)
m.geo_json(geo_path=topo_path, topojson="objects.object")
inline_map(m)
```
